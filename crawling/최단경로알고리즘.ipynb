```python
from itertools import *

position = [1, 2, 3, 4, 5]

dis = [[0, 17.4, 23.5, 6.3, 11.8],
       [14.2, 0, 34.3, 11.4, 5.5],
       [25.6, 38.0, 0, 26.2, 39.4],
       [5.0, 14.9, 27.7, 0, 9.3],
       [11.5, 5.6, 31.5, 12.0, 0]
       ]    


printlst = list(permutations(position, 5))
```

```python
print(printlst)
```

```python
aa = []
for i in range(len(printlst[0])-1):
    sum = 0
    for j in range(len(printlst[0])-1):
        a = dis[i][j]
        sum += a
    aa.append(sum)
print(aa)
```
