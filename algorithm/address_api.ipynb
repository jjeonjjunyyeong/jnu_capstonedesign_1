```python
li = []
routes = []
lst_len = len(lst)
# 장소에 대한 정보가 담겨있음
copy_lst = copy.deepcopy(lst)


# 일자 만큼 2차원 배열로 생성
for i in range(lst_len):
    li.append([])
    routes.append([])

for date in lst:
    lst_index = lst.index(date)
    date_len = len(date)
    li[lst_index] = [[0] * date_len for _ in range(date_len)]
    routes[lst_index] = [[0] * date_len for _ in range(date_len)]
# point to point에 대한 거리 및 시간 정보를 담음
copy_li = copy.deepcopy(li)

# routes 에는 알고리즘을 위한 각 거리만 담음 -> 유향 그래프
```

```python
for date in lst:
    lst_index = lst.index(date)
    for i in range(date_len):
        for j in range(date_len):
            print(date[j])
            route_info = gmaps.directions(date[i], date[j], mode='driving', units='metric', language='ko')
            # 거리(km)
            dist = route_info[0]['legs'][0]['distance']['text'][:-3]
            # 시간(분)
            time = route_info[0]['legs'][0]['duration']['text']
            # 출발지 주소
            start_addr = route_info[0]['legs'][0]['start_address']
            # 출발지 위도 경도
            start_lat = route_info[0]['legs'][0]['start_location']['lat']
            start_lon = route_info[0]['legs'][0]['start_location']['lng']
            # 도착지 주소
            end_addr = route_info[0]['legs'][0]['end_address']
            # 도착지 위도 경도
            end_lat = route_info[0]['legs'][0]['end_location']['lat']
            end_lon = route_info[0]['legs'][0]['end_location']['lng']
            if(i != j):
                copy_li[lst_index][i][j] = route_information(dist, time)
                routes[lst_index][i][j] = float(dist)
        copy_lst[lst_index][i] = places_information(start_addr, start_lat, start_lon, end_addr, end_lat, end_lon)
```

```python
# 모든 경우의 수 중 최단 거리를 담을 리스트 선언
shortest_route = []

cases = []
num_of_cases = []
for i in range(len(lst)):
    print(i)
    cases.append([])
    cases[i] = list(range(len(lst[i])))
    shortest_route.append([])
    num_of_cases.append(list(permutations(cases[i])))
```

```python
count = 0
for date in num_of_cases:
    num_of_cases_index = num_of_cases.index(date)
    
    print(num_of_cases_index)
    for i in date:
        rep = len(i) - 1
        cnt = 0.0
        for j in range(rep):
            cnt += routes[count][i[j]][i[j + 1]]
        shortest_route[count].append(cnt)
    count += 1
```

```python
np_shortest_route = np.array(shortest_route)
```

```python
# 가장 짧은 거리의 인덱스를 구함
for i in range(len(shortest_route)):
    min_index = np.argmin(np_shortest_route[i])
    print("-----------------------------------------------------")
    print(f"{i + 1} 일차")
    print("최솟값의 인덱스:", min_index)
    print("최단 경로 :", num_of_cases[i][min_index])
    print("최단 경로의 거리 :", np.round(shortest_route[i][min_index], 1), "km")
    print("-----------------------------------------------------")
```
